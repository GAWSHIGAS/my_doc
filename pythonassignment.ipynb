```python
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
vehicles = [42000,46000,48000,52000,55000,50000,57000,59000,58000,61000,63000,65000]

plt.bar(months, vehicles, color='skyblue')
plt.title("Monthly Vehicle Traffic Comparison")
plt.xlabel("Months")
plt.ylabel("Vehicles Passed")
plt.show()
```

```python
plt.pie(vehicles, labels=months, autopct='%1.1f%%', startangle=90)
plt.title("Monthly Contribution to Total Annual Traffic")
plt.show()
```

```python
accidents = [45,42,50,48,52,55,60,58,62,65,63,68]

plt.hist(accidents, bins=5, color='lightgreen', edgecolor='black')
plt.title("Distribution of Monthly Accidents")
plt.xlabel("Number of Accidents")
plt.ylabel("Frequency")
plt.show()
```

```python
fuel_price = [97,98,99,100,101,102,104,105,106,107,108,109]

plt.plot(fuel_price, vehicles, marker='o', linestyle='-', color='orange')
plt.title("Fuel Price vs Vehicle Traffic")
plt.xlabel("Fuel Price (₹/L)")
plt.ylabel("Vehicles Passed")
plt.show()
```
